```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
# Read in file
df = pd.read_csv("Resources/ABMatchData.csv")

# Separate year and add as new column
yr = df["Date"].str.split("-", 1)
year = [y[0] for y in yr]
df["Year"] = year
df
```

### Total match results vs. results over time (subplot: scatter, pie)

```python
# Total wins, losses, draws
df["Result"].value_counts()
```

```python
grouped_data = df.groupby(["Year", "Result"], as_index=False).size()
```

### Relationship between point gap and rating

```python
x = df["Point Gap"]
y = df["Rating"]

plt.figure(figsize=(10,3))
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Point Gap")
plt.ylabel("Rating")
plt.show()
```
